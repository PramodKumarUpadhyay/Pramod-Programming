```python
# Basic Packages #
import pandas as pd
import numpy as np
import statistics as st
import os as os
import seaborn as sns

# Visualization #
import matplotlib.pyplot as plt
```

```python
# Lets generate a random series of age of Iphone buyers
# np.random.seed(7)
series = np.random.normal(loc = 30, scale = 10, size = 1000)
# loc - Mean, Scale = Std Dev, Size = Output shape)
series
```

```python
np.random.seed(333888)
series = np.random.normal(loc = 30, scale = 10, size = 1000)
series
```

```python
np.mean(series)
```

```python
np.mean(series).round()
print ("Mean of the series is: ", np.mean(series).round())
```

```python
np.median(series).round()
```

```python
np.std(series).round()
print("Standard Deviation of teh series is: ", np.std(series).round())
```
