```python
# Lets generate a random series of age of Iphone buyers

np.random.seed(130)
series = np.random.normal(loc = 20, scale = 3, size=80)   # Loc - Mean  # scale = SD  # Size - Number of samples
```

```python
series[:5]     # to display first 5 numbers
```

```python
series = series.round(2)
```

```python
series[:5]
```

```python
print ("Mean of series: ", round(np.mean(series),0))              # series.mean()    # Mean of the series
print ("Median of series: ", round(np.median(series), 0))         # Median of the series
print ("Mean: {0}, Median: {1}".format(round(np.mean(series),0), round(np.median(series),0)))
print ("Standard Deviation: ", round(series.std(),0))             # np.std(series)   # Standard Deviation of data)
print ("Variance: ", round(series.var(),0))                       # np.var(series)   # Variance of data 
print ("Length/Number of elements in series: ", len(series))      # Length/Count/Number of rows in data
```

```python
plt.hist(series, color = "Green", alpha = 0.7)
plt.axvline(series.mean(), color = "red")
plt.axvline(series.mean() + series.std(), color = "red", ls = "--")
plt.axvline(series.mean() - series.std(), color = "red", ls = "--")
plt.axvline(series.mean() + 2* series.std(), color = "red", ls = "--")
plt.axvline(series.mean() - 2* series.std(), color = "red", ls = "--")
plt.axvline(series.mean() + 3* series.std(), color = "red", ls = "--")
plt.axvline(series.mean() - 3* series.std(), color = "red", ls = "--")
plt.xlabel("PLAYING CODE", fontsize = 12, color = "Blue")
plt.ylabel("COUNTRY", fontsize = 12, color = "Blue")
plt.title("Rolewsie Status",  fontsize = 12, Color = "Blue")
plt.grid()
plt.show()
```

```python
u = series.mean()
sd = series.std()
var= series.std()
```

```python
data = series[(series > u - sd) & (series < u + sd)]
data_1sd = len(data)*100/len(series)
print ("Percentage of data covering under 1 SD: {}%".format(data_1sd))
```

```python
data = series[(series > u - 2*sd) & (series < u + 2*sd)]
data_2sd = len(data)*100/len(series)
print ("Percentage of data covering under 2 SD: {}%", format(data_2sd))
```

```python
data = series[(series > u - 3*sd) & (series < u + 3*sd)]
data_3sd = len(data)*100/len(series)
print ("Percentage of data covering under 3 SD: {}%", format(data_3sd))
```
